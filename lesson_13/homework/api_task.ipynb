##### Subtask 4

```python
launches_df = pd.DataFrame(requests.get('https://api.spacexdata.com/v3/launches').json())
capsules_df = pd.DataFrame(requests.get('https://api.spacexdata.com/v3/capsules').json())

merged_df = launches_df.merge(capsules_df, left_on='launch_date_utc', right_on='original_launch', how='left')

merged_df
```

```python
dragon_one = merged_df[merged_df['type'] == 'Dragon 1.0']
dragon_one
```

```python
dragon_one_one = merged_df[merged_df['type'] == 'Dragon 1.1']
dragon_one_one
```

```python
print(f"Avg landings for Dragon 1.0: {dragon_one['landings'].mean()}.")
print(f"Avg landings for Dragon 1.1: {dragon_one_one['landings'].mean()}.")
```
